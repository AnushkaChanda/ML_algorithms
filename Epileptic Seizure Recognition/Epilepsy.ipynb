```python
#as data is imbalanced:
from imblearn.over_sampling import SMOTE

# Assuming 'X' should contain all columns except 'y' (the target variable) and 'Unnamed'
X = df.drop(['y', 'Unnamed'], axis=1)  # Assign values to X
smote=SMOTE(sampling_strategy='auto', random_state=42)
X_balanced,y_balanced=smote.fit_resample(X,df['y'])

# Check new class distribution
print(y_balanced.value_counts())
```

```python
#Feature Sampling

from sklearn.preprocessing import StandardScaler

# Standardize the feature values
scalar=StandardScaler()
X_scaled=scalar.fit_transform(X_balanced)

# Convert back to DataFrame
X_scaled=pd.DataFrame(X_scaled,columns=X.columns)

# Check transformed data
X_scaled.head()
```

```python
#Split Data for Training & Testing

from sklearn.model_selection import train_test_split

# 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced)

print("Data split successfully!")
```
